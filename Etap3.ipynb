W przypadku danych, które możemy podzielić liniowo na dwa zbiory zazwyczaj algorytm klasyfikacji próbuje znaleźć granicę pomiędzy zbiorami, tak by zminimalizować błąd niepoprawnej klasyfikacji.

##ISOLATION FOREST

Podejście to wykorzystuje fakt, że anomalie mają wartości zmienne bardzo różniące się od wartości normalnych. Algorytm działa świetnie z bardzo dużymi zbiorami danych wymiarowych

##ISOLATION FOREST - idea działania

Algorytm lasu izolacji wykorzystuje podstawowe właściwości anomalii:


*   jest ich niewiele i występują dość rzadko
*   są inne – mają wartości lub atrybuty mocno różne od pozostałych wartości

Rdzeniem algorytmu jest „izolowanie” anomalii poprzez tworzenie drzew decyzyjnych z losowymi atrybutami. Losowe partycjonowanie daje zauważalne krótsze ścieżki dla anomalii, ponieważ mniej wystąpień (anomalii) skutkuje mniejszymi partycjami.


Isolation Forest jest bardzo podobny do Random Forests i jest zbudowany w oparciu o zestaw drzew decyzyjnych dla danego zbioru danych. Istnieją jednak pewne różnice. Isolation Forest identyfikuje anomalie jako obserwacje o krótkich średnich długościach ścieżek na drzewach izolacyjnych. Dla każdego drzewa izolacyjnego stosowana jest procedura:

Wybierz losowo dwie funkcje.


*   Podziel punkty danych, losowo wybierając wartość między minimum a maksimum wybranych cech.
*   Podział obserwacji jest powtarzany rekurencyjnie, aż wszystkie obserwacje zostaną wyizolowane.

##Neural network

Autokoder to specjalny rodzaj sieci neuronowej, która jest wytrenowana do kopiowania danych wejściowych do danych wyjściowych. Autoenkoder uczy się kompresować dane, minimalizując jednocześnie błąd rekonstrukcji.

Wykrywanie anomali, opiera się o porównanie ustalonego progu z utratą rekonstrukcji.

Autoencoder:

````python
import tensorflow as tf

model = tf.keras.models.Sequential()

# encoder
model.add(
    tf.keras.layers.Dense(
        units=64, 
        input_dim=x_train.shape[1],
        activation=tf.nn.relu
    ))
model.add(
    tf.keras.layers.Dense(
        units=16,
        activation=tf.nn.relu
    ))
model.add(
    tf.keras.layers.Dense(
        units=4,
        activation=tf.nn.relu
    ))

# decoder
model.add(
    tf.keras.layers.Dense(
        units=16,
        activation=tf.nn.relu
    ))
model.add(
    tf.keras.layers.Dense(
        units=64,
        activation=tf.nn.relu
    ))
    
model.add(
    tf.keras.layers.Dense(
        units=x_train.shape[1],
    ))
````

**Losses**

Celem funkcji straty jest obliczenie wielkości, którą model powinien minimalizować podczas uczenia.

**MeanSquaredError**

Oblicza średnią kwadratów błędów między etykietami a prognozami.

**Optimizer**

Optymalizatory to algorytmy lub metody używane do zmiany atrybutów sieci neuronowej, takich jak wagi i szybkość uczenia się, w celu zmniejszenia strat.

**Adam**

Optymalizacja Adama to stochastyczna metoda opadania gradientu oparta na adaptacyjnej estymacji momentów pierwszego i drugiego rzędu.

````python
model.compile(loss='mse', optimizer='adam')
model.summary()

model.fit(x_train, x_train, epochs=32)
````

````python
from sklearn import metrics

test = model.predict(x_test)
diff = np.sqrt(metrics.mean_squared_error(test, x_test))
train_loss = tf.keras.losses.mae(test, x_test)

print('Błąd dopasowania: ', diff)

threshold = np.mean(train_loss) + np.std(train_loss)
print("Threshold: ", threshold)
````

###Neural Adam v2

````python
x=x.to_numpy()  #konwersja danych na numpy

#standaryzacja danych v2
m = np.mean(x, 0)
s = np.std(x, 0)
x = (x-m)/s
````

````python
#zaimplementowanie oraz wytrenowanie autoenkodera

#wyznaczenie bledu sredniokwadratowego
import jax
import jax.numpy as jnp
from jax.example_libraries.optimizers import adam
from jax.example_libraries import stax

def vector_mse(y,yhat):
    return jnp.mean(jnp.square(y-yhat), axis=-1)

#ustalenie odpowiednich parametrów
batch_size=128
steps=20000
learning_rate=0.001
train_split=int(0.6*x.shape[0])#podzial na wartosci treningowe
k = jax.random.PRNGKey(42) # klucz do liczb losowych


#konstruowanie encodera i decodera w zależności od ustalonych parametrow
encoder = stax.serial(
    stax.Dense(64),
    stax.Relu,
    stax.Dense(32),
    stax.Relu,
    stax.Dense(16),
    stax.Relu,
    stax.Dense(4)
)

decoder = stax.serial(
    stax.Dense(16),
    stax.Relu,
    stax.Dense(32),
    stax.Relu,
    stax.Dense(x.shape[1])
)

#stworzenie sieci neuronowej w oparciu o stax.serial
nn = stax.serial(encoder,decoder)
init_nn, apply_nn = nn
opt_init, opt_update, get_params, = adam(learning_rate)

@jax.jit
def step(i, opt_state, x, y):
    params=get_params(opt_state)
    #kara za zle dopasowanie przy uzyciu MSE 
    def loss(params, x, y):
        reconstruction = apply_nn(params,x)
        return jnp.mean(vector_mse(y,reconstruction))
    
    g = jax.grad(loss)(params, x, y)
    opt_state = opt_update(i, g, opt_state)
    step_loss = loss(get_params(opt_state),x,y)
    return opt_state, step_loss

@jax.jit
def test_loss(opt_state,x,y):
    params = get_params(opt_state)
    reconstruction = apply_nn(params,x)
    return jnp.mean(vector_mse(y,reconstruction)) #blad sredniokwadratowy

_,params = init_nn(k,(batch_size,x.shape[1]))
opt_state = opt_init(params)

losses = []
test_losses = []
train_steps = []

#wytrenowanie modelu
for i in range (steps):
    idx = np.random.randint(0,train_split,size=batch_size)
    bx = x[idx,...]
    
    opt_state, step_loss = step(i, opt_state,bx,bx)
    
    if i % 100 == 0:
        tx = x[train_split:,...]
        test_losses.append(test_loss(opt_state,tx,tx))
        train_steps.append(i)
        losses.append(test_loss(opt_state,x[:train_split,...],x[:train_split,...]))


````

````python

#wizualizacja
plt.plot(train_steps,losses,label='train')
plt.plot(train_steps,test_losses,label='test')

plt.legend()
print(test_losses[-1])
````
